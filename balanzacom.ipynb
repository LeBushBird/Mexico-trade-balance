```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
balanza = pd.read_csv('balanza.csv')
balanza
```

```python
total = balanza['Exportaciones'] - balanza['Importaciones']

x_axis = [i for i in range(0,365,12)]
labels = [j for j in range(1991,2022)]

plt.figure(figsize=(20, 10))
plt.bar(balanza['Periodos'], total)
plt.xticks(x_axis, labels)
plt.style.use('Solarize_Light2')
plt.show()
```
